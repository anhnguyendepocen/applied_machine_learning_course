### Standard Scalar
1. Mean
2. Standard Deviation

```python
import numpy as np
def mean(X):
    return np.sum(X)/len(X)
```

```python
age_mean = mean(df.age.tolist())
```

<img src="std.PNG">

```python
class StandardScalar():
    def fit(X):
        compute mean and standard deviation assing to self
    def tansform(X):
        returns (x-u)/s
```

```python
def standard_deviation(X,u):
    return np.sqrt(np.sum(np.array([(x-u)**2 for x in X])/len(X)))
```

```python
standard_deviation(df.age.tolist(),age_mean)
```
